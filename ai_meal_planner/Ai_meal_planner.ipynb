```python
df_cleaned = df.drop(columns=["Unnamed: 0"])
#converting time
def convert_time_to_minutes(time_str):
    if pd.isna(time_str):
        return None  # Keep NaN values
    time_str = str(time_str).lower()
    minutes = 0
    hours_match = re.search(r"(\d+)\s*hr", time_str)
    minutes_match = re.search(r"(\d+)\s*min", time_str)
    
    if hours_match:
        minutes += int(hours_match.group(1)) * 60
    if minutes_match:
        minutes += int(minutes_match.group(1))
    
    return minutes if minutes > 0 else None  # Convert invalid values to NaN

# Apply conversion to prep_time, cook_time, total_time
df_cleaned["prep_time"] = df_cleaned["prep_time"].apply(convert_time_to_minutes)
df_cleaned["cook_time"] = df_cleaned["cook_time"].apply(convert_time_to_minutes)
df_cleaned["total_time"] = df_cleaned["total_time"].apply(convert_time_to_minutes)

# Fill missing values for total_time if prep_time and cook_time are available
df_cleaned["total_time"] = df_cleaned.apply(
    lambda row: row["prep_time"] + row["cook_time"] if pd.isna(row["total_time"]) and 
    pd.notna(row["prep_time"]) and pd.notna(row["cook_time"]) else row["total_time"], axis=1
)
```

# For prep_time, cook_time, and total_time:

**If total_time is missing but prep_time and cook_time exist, we can fill total_time = prep_time + cook_time.**

**If cook_time is missing but total_time and prep_time exist, we can fill cook_time = total_time - prep_time.**

**If prep_time is missing but total_time and cook_time exist, we can fill prep_time = total_time - cook_time.**


```python
# Fill missing total_time where possible
df_cleaned["total_time"] = df_cleaned.apply(
    lambda row: row["prep_time"] + row["cook_time"] if pd.isna(row["total_time"]) and 
    pd.notna(row["prep_time"]) and pd.notna(row["cook_time"]) else row["total_time"], axis=1
)

# Fill missing cook_time where possible
df_cleaned["cook_time"] = df_cleaned.apply(
    lambda row: row["total_time"] - row["prep_time"] if pd.isna(row["cook_time"]) and 
    pd.notna(row["total_time"]) and pd.notna(row["prep_time"]) else row["cook_time"], axis=1
)

# Fill missing prep_time where possible
df_cleaned["prep_time"] = df_cleaned.apply(
    lambda row: row["total_time"] - row["cook_time"] if pd.isna(row["prep_time"]) and 
    pd.notna(row["total_time"]) and pd.notna(row["cook_time"]) else row["prep_time"], axis=1
)

# Function to extract numeric values from yield (e.g., "4 servings" -> 4)
def extract_yield(yield_str):
    if pd.isna(yield_str):
        return None
    match = re.search(r"(\d+)", str(yield_str))
    return int(match.group(1)) if match else None

# Apply extraction to yield column
df_cleaned["yield"] = df_cleaned["yield"].apply(extract_yield)

# Fill missing yield values with servings if available
df_cleaned["yield"] = df_cleaned["yield"].fillna(df_cleaned["servings"])

# Display updated null counts
df_cleaned.isnull().sum()
```

# Fill remaining missing time values with median of respective columns

```python
df_cleaned["prep_time"].fillna(df_cleaned["prep_time"].median(), inplace=True)
df_cleaned["cook_time"].fillna(df_cleaned["cook_time"].median(), inplace=True)
df_cleaned["total_time"].fillna(df_cleaned["total_time"].median(), inplace=True)

df_cleaned.isnull().sum()
```

# **Creating a new column in order to store the Calories and other nutritional values**

```python
import re

def extract_nutrient_value(nutrition_text, nutrient_name):
    match = re.search(rf"{nutrient_name} (\d+)g", str(nutrition_text))
    return int(match.group(1)) if match else 0
```

```python
df_cleaned["fat_g"] = df_cleaned["nutrition"].apply(lambda x: extract_nutrient_value(x, "Total Fat"))
df_cleaned["carbs_g"] = df_cleaned["nutrition"].apply(lambda x: extract_nutrient_value(x, "Total Carbohydrate"))
df_cleaned["protein_g"] = df_cleaned["nutrition"].apply(lambda x: extract_nutrient_value(x, "Protein"))

df_cleaned["estimated_calories"] = (df_cleaned["fat_g"] * 9) + (df_cleaned["carbs_g"] * 4) + (df_cleaned["protein_g"] * 4)

df_cleaned[["nutrition", "fat_g", "carbs_g", "protein_g", "estimated_calories"]].head(10)
```

# identifying outliers

```python
import matplotlib.pyplot as plt  
import seaborn as sns  

numerical_cols = ["fat_g", "carbs_g", "protein_g", "estimated_calories"]
plt.figure(figsize=(12, 6))
df_cleaned[numerical_cols].boxplot()
plt.xticks(rotation=45)
plt.title("Boxplot of Nutritional Values")
plt.show()
```

```python
Q1 = df_cleaned[numerical_cols].quantile(0.25)
Q3 = df_cleaned[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df_cleaned[numerical_cols] < (Q1 - 1.5 * IQR)) | 
            (df_cleaned[numerical_cols] > (Q3 + 1.5 * IQR)))

print("Number of outliers per column:")
print(outliers.sum())
```

```python
df_cleaned = df_cleaned[~((outliers).any(axis=1))]
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_cleaned[["fat_g", "carbs_g", "protein_g", "estimated_calories"]])
plt.title("Boxplot After Handling Outliers")
plt.show()
```

```python
def remove_outliers(df, columns):
    for col in columns:
        Q1, Q3 = df[col].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

columns_to_check = ["fat_g", "carbs_g", "protein_g", "estimated_calories"]
df_cleaned = remove_outliers(df_cleaned, columns_to_check)
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_cleaned[["fat_g", "carbs_g", "protein_g", "estimated_calories"]])
plt.title("Boxplot After Handling Outliers")
plt.show()
```

# Normalize the Features

```python
from sklearn.neighbors import KNeighborsRegressor
X = df_cleaned[["fat_g", "carbs_g", "protein_g", "estimated_calories"]]
knn_model = KNeighborsRegressor(n_neighbors=5, metric='euclidean')
knn_model.fit(X, X)
```

```python
import numpy as np

def recommend_recipes(user_fat, user_carbs, user_protein, user_calories, top_n=5):
    user_input = np.array([[user_fat, user_carbs, user_protein, user_calories]])  
    user_input_df = pd.DataFrame(user_input, columns=["fat_g", "carbs_g", "protein_g", "estimated_calories"])  

    distances, indices = knn_model.kneighbors(user_input_df, n_neighbors=top_n)

    return df_cleaned.iloc[indices[0]][["recipe_name", "fat_g", "carbs_g", "protein_g", "estimated_calories", "directions"]]
recommended_recipes = recommend_recipes(user_fat=50, user_carbs=150, user_protein=100, user_calories=2000)
print(recommended_recipes)
```
